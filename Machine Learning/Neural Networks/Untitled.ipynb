# Нейронные сети
Нейронная сеть - моедль, вдохновленная биологическими нейронами. 
Состоит из слоев нейронов, которые соединены между собой весами. Каждый нйрон принимает входные данные, применяет линейную комбинацию и функцию активации, затем предает результат дальше.

**Пример простейшей нейронной сети:**
* **Входной слой:** Признаки $x_{1}, x_{2}, ..., x_{k}$.
* **Скрытый слой:** Несколько нейронов, которые ообрабатывают входные данные.
* **Выходной слой:** Предсказание (напр. вероятность класса для классификации.
### Математичекая модель одного нейрона
Каждый нейрон выполняет следующие шаги:
1. Линейная комбинация входов:
    $$ z = w_{0} + w_{1}x_{1} + w_{2}x_{2} + ... + w_{k}x_{k} = w^{T}x $$
2. Применение функции активации $g(z)$:
   $$ a = g(z) $$
   * Для классификации часто используют сигмоиду (как в логистической регресси):
    $ g(z) = \frac{1}{1 + e^{-z}} $.
   * Для скрытых слоев нейронных сетях чаще используют ReLU:
     $ g(z) = max(0, z) $.
3. Структура нейронной сети
   Предположим, у нас есть сеть с одним скрытым слоем:
   * **Входной слой:** $ x = [x_1, x_2, ..., x_k] $.
   * **Скрытый слой:** $ h $ нейронов. Для каждого  нейрона j:
     $$ z^{[1]}_j = w^{[1]}_{j} * x + b^{1}_{j},   a^{[1]}_{j} = g(z^{[1]}_{j}) $$
     $w^{[1]}_{j}$ - веса, $b^{[1]}_{j}$ - смещение, $g$ - функция активации
   * **Выходной слой** Один нейрон (для бинарной классификации):
     $$ z^[2] = w^{[2]} * a^{[1]} + b^{[2]}, \hat{y} = \sigma(z^{[2]})$$
     $\sigma$ - сигмоида, $\hat{y}$ - предсказанная вероятность класса 1.
